### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
## Nomeando cada loja
lojas = [loja, loja2, loja3, loja4]
loja_nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

## Exibindo o faturamento total com casas decimais em reais
for i, df in enumerate(lojas):
  print(f'O faturamento da {loja_nomes[i]} foi de R${df["Preço"].sum():,.2f}\n')

## Transofmando em graficos
plt.bar(loja_nomes, [df['Preço'].sum() for df in lojas])
plt.ylabel('Faturamento')
plt.title('Faturamento por Loja')
plt.show()
```

```python
import matplotlib.pyplot as plt
```

# 2. Vendas por Categoria


```python
## Somando vendas por categoria
for i, df in enumerate(lojas):
    print(f'\n Produtos mais vendidos por categoria da {loja_nomes[i]} \n')
    print(df['Categoria do Produto'].value_counts().to_string())
```

# 3. Média de Avaliação das Lojas

```python
## Somando a media de avaliações de cada loja

for i, df in enumerate(lojas):
    media_das_avaliacoes = df['Avaliação da compra'].mean()
    print(f'A média de avaliação da {loja_nomes[i]} foi de {media_das_avaliacoes:,.2f} \n')

## Transformando em graficos

plt.plot(loja_nomes, [df['Avaliação da compra'].mean() for df in lojas])
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
## Buscando os produtos mais vendidos

for i, df in enumerate(lojas):
    print(f'\n Produtos mais vendidos da {loja_nomes[i]}')
    print(df['Produto'].value_counts().head(3).to_string())
## Buscando os menos vendidos
    print(f'\n Produtos menos vendidos da {loja_nomes[i]}')
    print(df['Produto'].value_counts().tail(3).to_string())

## Gerando graficos dos resultados

for i, df in enumerate(lojas):
  plt.title(f'Produtos mais e menos vendidos da {loja_nomes[i]} \n')
  plt.bar(df['Produto'].value_counts().head(3).index, df['Produto'].value_counts().head(3))
  plt.bar(df['Produto'].value_counts().tail(3).index, df['Produto'].value_counts().tail(3))
  plt.tight_layout()
  plt.xticks(rotation=45)
  plt.show()
```

# 5. Frete Médio por Loja


```python
## Garantindo que exista apenas um nome para cada media de frete
fretes_medios = []
nomes_lojas = []

## Calculando o frete
for i, df in enumerate(lojas):
    fretes_medios.append(df['Frete'].mean())
    nomes_lojas.append(loja_nomes[i])
## Exibindo resultados
    print(f'O frete médio da {loja_nomes[i]} foi de R${df["Frete"].mean():,.2f} \n')

## Gerando grafico
plt.figure(figsize=(6, 6))
plt.pie(fretes_medios, labels=nomes_lojas, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição do Frete Médio por Loja')
plt.show()
```
